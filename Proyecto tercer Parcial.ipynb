# Flujos compresibles

Un flujo compresible es aquel que cuando cambia la presión varía la densidad. También se puede medir la compresibilidad mediante la velocidad en que se transmiten pequeñas pertubaciones dentro del mismo fluido, esta velocidad se le llama velocidad sónica y este tipo de flujos poseen bajas velocidades sónicas. 

##### Número de Mach
La velocidad del sonido en el agua con 20°C es de 1483.2 $\frac{m}{s}$ y en el aire es de 331.3$\frac{m}{s}$. El número de Mach es la relación entre la velocidad del flujo (V) y la velocidad del sonido (c).

$M=\frac{M}{c}$

Y clásifica los flujos de la siguiente manera

- Flujo incompresible: Mach menor que 0.3.
- Flujo subsónico: Mach entre 0.3 y 0.8.
- Flujo transónico : Mach entre 0.8 y 1.2
- Flujo supersónico: Mach enre 1.2 y 3.0
- Flujo hipersónico. Mach mayor que 3.0

(Flujo Compresible | la Guía de Química, s. f.-b)

La densidad de un fluido se define como la relación entre la masa y el volumen con unidades $\frac{kg}{m^3}$

$\rho=\frac{m}{v}$

La energía interna de un sistema es aquella que mide el movimiento de los átomos y moléculas que componen al sistema y la unidad de medida que se usan son los Joules Cristhian et al. (2021).

### Problema a resolver

Se propone resolver un sistema de ecuaciones diferenciales acoplado con el fin de observar el cambio de la densidad, presión y velocidad a lo largo del flujo. El flujo es un gas ideal, unidimensonal, estacionario y con un número de Mach supersónico. No existe intercambio de calor y las variaciones en la dirección transversal son despreciables. 

El sistema de ecuaciones diferenciales es el siguiente:

$\frac{\partial\rho}{\partial t}+\frac{\partial \rho u }{\partial x}=0$

$\frac{\partial\rho u}{\partial t}+\frac{\partial(\rho u^2+ P)}{\partial x}=0$

$\frac{\partial \rho E}{\partial t}+\frac{\partial (\rho u E + Pu)}{\partial x}=0$

Donde 
- $E= e+\frac{u^2}{2}$ y e es la energía interna específica del gas
- $\rho$ es la densidad
-P es la presión 
- u es la velocidad del gas 



Se tienen 4 variables sin embargo, la energía esta en términos de u, por lo que se reduce a tres variables y tres ecuaciones

El sistema se puede reescribir de la siguiente forma:

$\frac{\partial \rho}{\partial t}= -\frac{\partial \rho u}{\partial x}$

$\frac{\partial \rho u}{\partial t} = - \frac{\partial (\rho u^2 + P)}{\partial x}$

$\frac{\partial \rho E}{\partial t} = -\frac{\partial (\rho u E + Pu)}{\partial x}$

Sustancia alcohol, densidad 780$\frac{kg}{m^3}$

```python
import numpy as np
import sympy as sy 
import matplotlib.pyplot as plt
```

```python
ti=0
tf=10
xi=0
xf=10
nx = 10
nt = 10
ht= (tf-ti)/(nt-1)
hx= (xf-xi)/(nx-1)
e=5
w=1
rho=np.zeros((nx,nt))
u=np.zeros((nx,nt))
E=np.zeros((nx,nt))
P=np.zeros((nx,nt))

def x(i):
    return xi + i*hx

def t(i):
    return ti + i*ht
```

```python
for i in range(nx):
    for j in range(nt):
        if j==0:
            rho[i,j]=x(i)+780
        else:
            rho[i,j]=0
print (rho)
```

```python
for i in range(nx):
    for j in range(nt):
        if j == 0:
            u[i, j] =x(i)+ 1/rho[i,j]
        else:
            u[i, j] = 0

print(u)
```

```python
for i in range(nx):
    for j in range (nt):
        if j==0:
            E[i,j] = e + ((u[i,j])**2)/2
        else:
            E[i,j]=0
print(E)
```

Suponiendo que P y V tiene una relación de $P=V^w$, donde w puede ser cualquier potencia, poniendo todo en términos de densidad queda que
$P=(\frac{m}{\rho})^w$

```python
for i in range(nx):
    for j in range(nt):
        if j==0:
            P[i,j]=(rho[i,j])**w
        else:
            P[i,j]=0
print (P)
```

```python
time_steps = np.arange(ti, tf) * ht
x_positions = np.arange(xi, xf) * hx

# Crear subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 15))

# Graficar rho
for i in range(xi, xf):
    axs[0].plot(time_steps, rho[i, :], label=f'Posición x={i}')
axs[0].set_title('Densidad (rho) a lo largo del tiempo')
axs[0].set_xlabel('Tiempo')
axs[0].set_ylabel('Densidad')
axs[0].legend()

# Graficar u
for i in range(xi, xf):
    axs[1].plot(time_steps, u[i, :], label=f'Posición x={i}')
axs[1].set_title('Velocidad (u) a lo largo del tiempo')
axs[1].set_xlabel('Tiempo')
axs[1].set_ylabel('Velocidad')
axs[1].legend()

# Graficar E
for i in range(xi, xf):
    axs[2].plot(time_steps, E[i, :], label=f'Posición x={i}')
axs[2].set_title('Energía (E) a lo largo del tiempo')
axs[2].set_xlabel('Tiempo')
axs[2].set_ylabel('Energía')
axs[2].legend()

# Graficar P
for i in range(xi, xf):
    axs[3].plot(time_steps, P[i, :], label=f'Posición x={i}')
axs[3].set_title('Presión (P) a lo largo del tiempo')
axs[3].set_xlabel('Tiempo')
axs[3].set_ylabel('Presión')
axs[3].legend()

plt.tight_layout()
plt.show()
```
