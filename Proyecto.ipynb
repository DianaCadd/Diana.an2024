# ECUACIÓN DE COLEBROOK-WHITE

#### Es un módelo matemático el cual permite cálcular la disminución de presión en la tuberias y permitir que sea el flujo sea eficiente 

$\frac{1}{\sqrt{f}}+2log[\frac{e/D}{3.7}+\frac{2.51}{Re\sqrt{f}}]=0$

Donde f es el coeficiente de fricción; 
e es Rugosidad absoluta;
D es el diametro de la computadora; 
Re es el número de Reynolds 
En esta función la variable que hay que encontrar es la f, y como se observa es una ecuación no lineal 


```python
import numpy as np
import sympy as sy 
import matplotlib.pyplot as plt
```

```python
Re=5000
D=0.5
e=3*10**-7
```

El número de Reynolds es de tipo turbulento, esto quiere decir que el movimiento del líquido es desordenado. Mientras que la rugosidad e es de un plástico y esta medido en metros

```python
def colwhite(f):
    return 1/f**(1/2)+np.log(((e/D)/3.7)+(2.51/(Re*(f**(1/2)))))
    
```

```python
def devcol(f):
    return (-1/(2*f**(3/2)))+2*(((-2.5/(2*Re))*(1/(f**(3/2))))/(((e/D)/3.7)+(2.5/(Re*(f**(1/2))))))*np.log(np.e)
```

```python
def newton(f):
    return f-(colwhite(f)/devcol(f))
```

```python
f=0.01
i=0
while f!=newton(f):
    f=newton(f)
    i=i+1
    print (f)
print (i)
    
```

## Gráfica de la función y como va convergiendo la raíz 

```python
x=np.linspace(0.001,0.05)
f1=colwhite(x)
f2=newton(x)
plt.plot(x,f1)
plt.plot(x,f2)
plt.legend(['colwhite','newton'])
plt.show()
```

## Gráfica de la función, la derivada y la convergencia de la raíz

```python
x=np.linspace(0.001,0.09)
f1=colwhite(x)
df1=devcol(x)
f2=newton(x)
plt.plot(x,f1)
plt.plot(x,df1)
plt.plot(x,f2)
plt.legend(['colwhite','decolwhite','newton'])
plt.show()
```

## Análisis de resultados 

Es recomendable que el punto donde se inicie las iteraciones sea de orden a la -2, ya que si se inicia con un número más grande la función Newton no converge y causa un error catastrófico. El error aproximado es pequeño por lo que puede ser una raiz "decente" para los datos ingresados.
